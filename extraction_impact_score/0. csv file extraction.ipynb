```python
import os
import json
import pandas as pd
```

Query: Investment

```python
json_file = 'investment_mean_output_cl3.5.json'
with open(json_file, 'r') as f:
    data = json.load(f)
```

```python
data
```

```python
# make dataframe from json
investment_df = pd.DataFrame(data)
cols = {'price_impact':'yt_price impact_investment', 'demand_impact':'yt_demand impact_investment', 'supply_impact':'yt_supply impact_investment'}
investment_df = investment_df.transpose()

investment_df.rename(columns=cols, inplace=True)
```

```python
investment_df.tail(30)
```

Supply Chain

```python
json_file = 'supply_chain_mean_output_cl3.5.json'
with open(json_file, 'r') as f:
    data = json.load(f)
```

```python
# make dataframe from json
supply_chain_df = pd.DataFrame(data)
cols = {'price_impact':'yt_price impact_supply_chain', 'demand_impact':'yt_demand impact_supply_chain', 'supply_impact':'yt_supply impact_supply_chain'}
supply_chain_df = supply_chain_df.transpose()

supply_chain_df.rename(columns=cols, inplace=True)
```

```python
supply_chain_df.tail(30)
```

```python
# save csv
investment_df.to_csv('IS_investment_3.5.csv')
supply_chain_df.to_csv('IS_supply_chain_3.5.csv')
```

```python
# merge investment_df and supply_chain_df
# bring index from investment_df disregarding the index of supply chain df
merged_df = pd.merge(investment_df, supply_chain_df, left_index=True, right_index=True)
```

```python
merged_df.tail(30)
```

```python
# save csv 
merged_df.to_csv('yt_impact score_240712.csv')
```
